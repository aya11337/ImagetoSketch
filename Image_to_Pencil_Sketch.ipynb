```python
import cv2
from google.colab.patches import cv2_imshow

# Step 1: Read the image
image = cv2.imread('/content/dog.jpg')
cv2_imshow(image)
```

```python
# Step 2: Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Invert the grayscale image
inverted_gray = 255 - gray_image

# Step 4: Blur the inverted image
blurred = cv2.GaussianBlur(inverted_gray, (15, 15), 0)

# Step 5: Invert the blurred image
inverted_blur = 255 - blurred

# Step 6: Create the pencil sketch by blending
pencil_sketch = cv2.divide(gray_image, 255 - blurred, scale=220.0)


# Display the original and the sketch
#cv2.imshow("Original Image", image)
cv2_imshow(pencil_sketch)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
